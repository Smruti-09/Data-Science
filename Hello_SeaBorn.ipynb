```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

var = [1,2,3,4,5,6,7]
var_1 = [2,3,6,9,11,15,24]
x_1 = pd.DataFrame({"var" : var, "var_1" : var_1})
sns.lineplot(x = "var", y = "var_1",data = x_1)
plt.show()
```

```python
plt.plot(var,var_1)
plt.show()
```
