# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data =pd.read_csv(r"D:\Data_Science_CSVs\tips.csv")
data
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data,hue = "sex")
plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue = "sex",palette="bwr")
plt.show()
```

```python
sns.pairplot(data,hue = "sex",palette="bwr",x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(data,hue = "sex",palette="bwr",kind="reg")
plt.show()
```

```python
sns.pairplot(data,hue = "sex",palette="bwr",kind="kde")
plt.show()
```

```python
sns.pairplot(data,hue = "sex",kind="kde",diag_kind="hist")
plt.show()
```
