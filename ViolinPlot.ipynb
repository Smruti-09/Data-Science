```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data =pd.read_csv(r"D:\Data_Science_CSVs\tips.csv")
data
```

```python
sns.violinplot(x="day",y = "total_bill", data =data, hue ="time",linewidth=1,palette= "Dark2",saturation=1,)
plt.show()
```

```python
sns.violinplot(x="time",y = "total_bill", data =data, order =["Dinner","Lunch"],palette ="icefire",saturation=0.5)
plt.show()
```

```python
sns.violinplot(x="total_bill",y = "day", data =data,hue="sex",split=True,scale="area",)
plt.show()
```

```python
sns.violinplot(x="day",y = "total_bill", data =data,hue="sex",split=True,scale="area",)
plt.show()
```

```python
sns.violinplot(x="day",y = "total_bill", data =data,hue="sex",split=True,scale="area",inner="quart")
plt.show()
```

```python
sns.violinplot(x=data["total_bill"])
plt.show()
```
