# Histogram Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset('penguins')
var
```

```python
sns.displot(var["bill_depth_mm"],kde = True,rug = True)
plt.show()
```

```python
sns.displot(var["bill_depth_mm"],bins = [17,18,19,20,21,22,23,24]) #bins = [170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],kde = True,rug = True,color='purple',log_scale = True)
plt.show()
```
