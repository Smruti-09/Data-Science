# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Data_Science_CSVs\tips.csv")
df
```

```python
sns.set(style="whitegrid")

sns.boxplot(x = "day", y = "total_bill",data = df,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x = "day", y = "total_bill",data = df,hue="time",showmeans = True,meanprops={"marker":"+"},palette="rocket")
plt.show()
```
