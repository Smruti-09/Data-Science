```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Data_Science_CSVs\tips.csv")
df
```

```python
fg = sns.FacetGrid(df)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg = sns.FacetGrid(df,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(df,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(df,col="sex",hue="day",palette="cool")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
